# Max Pooling

Reducción Dimensional

```python
import cv2
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
```

Cargamos la imágen deseada y cambiamos el tamaño.

```python
im = cv2.imread('street.jpg')
im = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
im = cv2.resize(im,None,fx=0.2,fy=0.2)
plt.imshow(im,cmap="gray")

# Image dimensions
im_h = im.shape[0]
im_w = im.shape[1]
```

```python
im = im.reshape(1,im_h,im_w,1)
```

Aplicamos la operación de pooling.

```python
pool = tf.nn.max_pool(value=im,ksize=[1,2,2,1],strides=[1,2,2,1],padding='SAME')
```

Creamos la sesión e inicializamos cualquier variable.

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

Solicitamos el resultado de la opreación de pooling.

```python
resultado = sess.run(pool)

plt.imshow(resultado[0,:,:,0],cmap="gray")
```
